```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/MLDatasets/cc_approvals.data', header=None)

# Separate the independent variables (X) and the dependent variable (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Identify categorical columns (assuming non-numeric data are categorical)
categorical_columns = X.select_dtypes(include=['object']).columns

# Apply one-hot encoding to categorical columns
X_encoded = pd.get_dummies(X, columns=categorical_columns)

# Ensure all column names are strings
X_encoded.columns = X_encoded.columns.astype(str)

# Display the first few rows of the encoded independent variables
print(X_encoded.head())

# Visualize the data using a scatter plot of the first two columns of independent variables
plt.scatter(X_encoded.iloc[:, 0], X_encoded.iloc[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter plot of the encoded independent variables')
plt.show()

# Isolation Forest for anomaly detection
# Initialize Isolation Forest with a contamination rate of 20%
clf = IsolationForest(contamination=0.2)

# Fit the Isolation Forest model to the encoded independent variables
clf.fit(X_encoded)

# Predict anomalies (1 for normal, -1 for anomaly)
predictions = clf.predict(X_encoded)

# Display the predictions
print(predictions)

# Find indices of anomalies
index = np.where(predictions < 0)

# Convert independent variables to numpy array for easy indexing
x_values = X_encoded.values

# Visualize the data with anomalies highlighted using Isolation Forest
plt.scatter(X_encoded.iloc[:, 0], X_encoded.iloc[:, 1])
plt.scatter(x_values[index, 0], x_values[index, 1], edgecolors="r", label='Anomalies')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Isolation Forest - Anomalies')
plt.legend()
plt.show()

# Gaussian model for anomaly detection
# Initialize Gaussian model with a contamination rate of 20%
gaussian_model = EllipticEnvelope(contamination=0.2)

# Fit the Gaussian model to the encoded independent variables
gaussian_model.fit(X_encoded)

# Predict anomalies (1 for normal, -1 for anomaly)
gaussian_predictions = gaussian_model.predict(X_encoded)

# Display the predictions
print(gaussian_predictions)

# Find indices of anomalies using Gaussian model
gaussian_index = np.where(gaussian_predictions < 0)

# Visualize the data with anomalies highlighted using Gaussian model
plt.scatter(X_encoded.iloc[:, 0], X_encoded.iloc[:, 1])
plt.scatter(x_values[gaussian_index, 0], x_values[gaussian_index, 1], edgecolors="r", label='Anomalies')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Gaussian Model - Anomalies')
plt.legend()
plt.show()
```
